# Polynomial Regression

```python
def generateX(N):
    x = np.random.random(N)*2*np.pi
    yd = np.sin(x)
    return x, yd
```

```python
x, y = generateX(200)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
def createA(x, D):
    A = np.ones(D+1)
    for d in range(1,  D+1):
        A[d]  = np.power(x, d)/np.power(2*np.pi, d)
    return A
```

```python
def init_w(D):
    w = np.random.random(D+1)
    return w
```

```python
def predict_y(w, x):
    D = len(w)-1
    A = createA(x, D)
    #print("A = ",A)
    #print("w = ", w)
    yhat = A.dot(w)
    return yhat
```

```python
def compute_se(yhat, y):
    se = (y-yhat)**2
    return se
```

```python
def calc_dw(x, yhat, y, D):
    dw = np.zeros(D+1)
    com = (-2)*(y-yhat)
    dw[0] += com
    for d in range(1, D+1):
        dw[d] +=com*np.power(x, d)/np.power(2*np.pi, d)
    return dw
```

```python
def update_w(w, dw, eta):
    w = w-dw*eta
    return w
```

```python
def train_w(x, y, w, Nepoch, eta):
    N = len(x)
    D = len(w)-1
    
    for epoch in range(Nepoch):
        for i in range(len(x)):
            #print(w, x[i])
            yhat = predict_y(w, x[i])
            #print(y[i], yhat)
            dw = calc_dw(x[i], yhat,  y[i], D)
            
            w = update_w(w, dw, eta)
    return w
```

```python
N = 300
D = 12
Nepoch = 300

eta = 0.001
se = 0
y_pred = []
w = init_w(D)
```

```python
x, y = generateX(N)
w =  train_w(x, y, w, Nepoch, eta)
```

```python
for i in range(N):
    y_pred.append(predict_y(w, x[i]))
    se += compute_se(y[i], y_pred[i])
    
mse = se/N
y_pred = np.array(y_pred)
```

```python
mse
```

```python
def plotmodel(x, y, yd):
    i = x.argsort()
    plt.figure()
    plt.plot(x[i], y[i], "g-o")
    plt.plot(x[i], yd[i],  "r-o")
    plt.ylabel("F(X)")
    plt.xlabel("X")
    plt.legend(["estimated", "True"])
    plt.title("Comparision")
    plt.show()
```

```python
plotmodel(x, y_pred, y)
```
